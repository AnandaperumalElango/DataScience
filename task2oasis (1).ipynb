```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import datetime as dt
import warnings

data = pd.read_csv("Unemployment in India.csv")
print(data.head())
```

```python
print(data.isnull().sum())
```

```python
data.describe().transpose()
```

```python
data.columns=["Region","Date","Frequency","Estimated Unemployment Rate (%)","Estimated Employed","Estimated Labour Participation Rate (%)","Area"]
#correlation
plt.style.use('seaborn-whitegrid')
plt.figure(figsize=(12,10))
sns.heatmap(data.corr())
plt.show()
```

```python
#data visualization
data.columns=["Region","Date","Frequency","Estimated Unemployment Rate (%)","Estimated Employed","Estimated Labour Participation Rate (%)","Area"]

plt.figure(figsize=(12,15))
plt.title("employment wrt to Region")
sns.histplot(x="Estimated Employed",hue="Region",data=data)
plt.show()
```

```python
plt.figure(figsize=(12,10))
plt.title("Indian Unemployment with respect to region")
sns.histplot(x="Estimated Unemployment Rate (%)",hue="Region",data=data)
plt.show()
```
